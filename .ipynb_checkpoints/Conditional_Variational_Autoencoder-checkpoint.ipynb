### Define Conditional Variational Autoencoder ###

```python
import math

def convert_labels_to_one_hot(c, num_classes):
    one_hot = torch.zeros((len(c), num_classes))  # initialize one-hot tensor with zeros
    for i, label in enumerate(c):
        one_hot[i, label] = 1
    return one_hot
def is_divisible_by(num, divisor):
    return int(num/divisor)*divisor == num

first_out_channels = 32
second_out_channels=64
third_out_channels = 128
fourth_out_channels=256
class MultiLabel_cVAE(nn.Module):
    def __init__(self, width, height, h_dim, z_dim, num_classes, input_channels, needs_one_hot=True, ):
        super(MultiLabel_cVAE, self).__init__()
        NUM_MAX_POOL = 3
        self.required_divisor = 2**NUM_MAX_POOL
        
        self.num_classes = num_classes
        self.width = width
        self.height = height
        self.z_dim = z_dim
        
        if(not is_divisible_by(self.height, self.required_divisor) or not is_divisible_by(self.width, self.required_divisor)):
            print("WARNING. Height (" + str(self.height)+ ") or Width (" + str(self.width)+ ") is not divisible by required divisor (" + str(self.required_divisor) + ")")
        self.size_of_flattened = int(fourth_out_channels * (self.height/self.required_divisor) * (self.width/self.required_divisor))
        # there are 2 max poolings which each half the sqrt dim of x.dim

        self.needs_one_hot = needs_one_hot
                
        self.encoderLayersCNN = nn.Sequential(
            nn.Conv2d(in_channels=input_channels, out_channels=first_out_channels, kernel_size=6, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=4, stride=2),
            nn.Conv2d(in_channels=first_out_channels, out_channels=second_out_channels, kernel_size=4, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=3, stride=2),
            nn.Conv2d(in_channels=second_out_channels, out_channels=third_out_channels, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=3, stride=2),
            nn.Conv2d(in_channels=third_out_channels, out_channels=fourth_out_channels, kernel_size=2, padding=1),
            nn.ReLU(),
            nn.Flatten()
        )
       
        self.encoderLayersLinear = nn.Sequential(
            nn.Linear(in_features=self.size_of_flattened + self.num_classes, out_features=h_dim),
            nn.ReLU())
        

        self.mu_sample_layer = nn.Linear(in_features=h_dim, out_features=z_dim)
        self.log_var_sample_layer = nn.Linear(in_features=h_dim, out_features=z_dim)
        
        # decoder part
        self.decoderLayersLinear = nn.Sequential(
            nn.Linear(z_dim+self.num_classes, h_dim),
            nn.ReLU(),
            nn.Linear(h_dim, self.size_of_flattened)
        )

        self.decoderLayersTransposeCNN = nn.Sequential(
            nn.ReLU(),
            nn.ConvTranspose2d(in_channels=fourth_out_channels, out_channels=third_out_channels, kernel_size=2, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(in_channels=third_out_channels, out_channels=second_out_channels, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(in_channels=second_out_channels, out_channels=first_out_channels, kernel_size=4, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(in_channels=first_out_channels, out_channels=input_channels, kernel_size=7, stride=2, padding=1, output_padding=1),
            nn.Sigmoid()
        )
        
    def encoder(self, x, c_one_hot):
        h = self.encoderLayersCNN(x)
        concat = torch.cat([h, c_one_hot], dim=1)
        h = self.encoderLayersLinear(concat)
        return self.mu_sample_layer(h), self.log_var_sample_layer(h) # mu, log_var
    
    def sampling(self, mu, log_var):
        std = torch.exp(0.5*log_var)
        eps = torch.randn_like(std)
        return eps.mul(std).add_(mu) # return z sample
        
    def decoder(self, z, c_one_hot):
        zc = torch.cat([z, c_one_hot], dim=1)
        x = self.decoderLayersLinear(zc)
        print(x.size())
        
        x = x.view(-1, fourth_out_channels, self.height/self.required_divisor, self.width/self.required_divisor)
        x = self.decoderLayersTransposeCNN(x)
        return x
    
    def forward(self, x, c):
        c_onehot = c
        if self.needs_one_hot:
            c_onehot = torch.nn.functional.one_hot(c, num_classes=self.num_classes)
        # flatten
        mu, log_var = self.encoder(x, c_onehot)
        z = self.sampling(mu, log_var)
        return self.decoder(z, c_onehot), mu, log_var
    
    def generate(self, c):
        z = torch.randn(len(c), self.z_dim).to(device)
        c_onehot = c
        if self.needs_one_hot:
            c_onehot = torch.nn.functional.one_hot(c, num_classes=self.num_classes).to(device)
        return self.decoder(z, c_onehot)
```

### Define Loss ###

```python
# return reconstruction error + KL divergence losses
def loss_function(recon_x, x, mu, log_var):
    #print the shapes
    BCE = F.binary_cross_entropy(recon_x, x, reduction='sum')
    KLD = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())
    return BCE + .15 * KLD
```

### Define Train and Test Methods ###

```python
def train(model, epoch, train_loader, optimizer, show_faces=False):
    model.train()
    train_loss = 0
    for batch_idx, (data, labels) in enumerate(train_loader):
        data = data.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        recon_batch, mu, log_var = model(data, labels)
        loss = loss_function(recon_batch, data, mu, log_var)
        
        loss.backward()
        train_loss += loss.item()
        optimizer.step()
        
        if batch_idx % 10 == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item() / len(data)))
        if batch_idx%30 ==0:
            plot_images(generate_images(model, 1))
            model.train()
    print('====> Epoch: {} Average loss: {:.4f}'.format(epoch, train_loss / len(train_loader.dataset)))

def test(model, test_loader):
    model.eval()
    test_loss= 0
    with torch.no_grad():
        for data, labels in test_loader:
            data = data.to(device)
            labels = labels.to(device)
            recon_batch, mu, log_var = model(data, labels)
            
            # sum up batch loss
            test_loss += loss_function(recon_batch, data, mu, log_var).item()
        
    test_loss /= len(test_loader.dataset)
    print('====> Test set loss: {:.4f}'.format(test_loss))
dataset = 'dataset'
```

### Define Sample Generation Methods ###

```python
import random
def generate_images(model, num_images, tgt_attributes=None):
    """
    Generates num_images images from the given model with the specified attribute labels.

    Args:
        model (MultiLabel_cVAE): The model to generate images from.
        num_images (int): The number of images to generate.
        attributes (dict): A dictionary of attribute labels to condition the generated images on. The keys of the dictionary
                           should be the attribute names (strings) and the values should be binary values indicating the
                           presence (1) or absence (0) of the attribute. If an attribute value is set to None, a random
                           value is generated for that attribute.
        celeba_dataset (CelebA): The CelebA dataset used to extract the attribute names. If not provided, the attribute
                                 names are assumed to be the standard CelebA attribute names.

    Returns:
        A tensor of shape (num_images, 3, H, W) representing the generated images.
    """
    # Load the attribute names from the CelebA dataset if not provided

    # Create a tensor to hold the generated images
    
    images = torch.zeros((num_images, 3, model.width, model.height))
    og_tgt_attributes = None if tgt_attributes is None else tgt_attributes.copy()
    # Generate the specified number of images
    for i in range(num_images):
        # Generate a random value for each attribute that is None
        if og_tgt_attributes is not None:
            for attr in attributes:
                if og_tgt_attributes[attr] is None:
                    tgt_attributes[attr] = random.randint(0, 1)

        # Convert the attributes dictionary to a tensor
        if tgt_attributes is None:
            c = torch.tensor([random.randint(0, 1) for _ in attributes]).unsqueeze(0)
        else:
            c = torch.tensor([tgt_attributes[attr] for attr in tgt_attributes]).unsqueeze(0)
        
        c = c.to(device)
        print("generating images with attributes", c)
        # Generate an image with the specified attribute labels
        image = model.generate(c)

        # Add the image to the tensor
        images[i] = image.squeeze()

    return images
def plot_images(images):
    celeb_vae.eval()
    image_count = len(images)
    nrows = 1 if image_count == 1 else 2
    ncols = (math.ceil(image_count/nrows))
    
    # Display the generated images
    import matplotlib.pyplot as plt
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(image_count, 4))
    for i in range(image_count):
        ax = axes[i//ncols, i%ncols] if nrows != 1 else axes
        ax.imshow(images[i].detach().permute(1, 2, 0))
        ax.axis('off')
    plt.show()
```
